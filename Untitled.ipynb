```python
base_url = 'https://www.kohls.com'
boots_links = []
for link in soup.find_all('a'):
    url = link.get('href')
    if url and '/catalog/' in url and '/catalog/shoes/' not in url:
        boots_links.append(url)
print(len(boots_links))        
        
```

```python
r = requests.get(base_url + boots_links[17])
soup = BeautifulSoup(r.text)
print(soup.title.string)
```
